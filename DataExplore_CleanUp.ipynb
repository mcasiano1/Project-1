```python
# Dependencies
import pandas as pd
import json
import numpy as np
```

```python
# The path to our CSV file
csv_path = "Resources/country_vaccinations.csv"

# Read our COVID data into pandas
covid_df = pd.read_csv(csv_path)
covid_df.head()
```

```python
#Remove NAN from DataFrame
covid_df=covid_df.fillna(0)
covid_df
```

```python
#Remove all decimals from the DataFrame
covid_df = covid_df.astype({"total_vaccinations":"int", "people_vaccinated": "int", "people_fully_vaccinated": "int", "daily_vaccinations_raw": "int", "daily_vaccinations": "int", "total_vaccinations_per_hundred": "int", "people_vaccinated_per_hundred": "int", "people_fully_vaccinated_per_hundred": "int", "daily_vaccinations_per_million": "int"})
covid_df
```

```python
#Delete "Source Website" from dataset
del covid_df["source_website"]
covid_df.head()
```

```python
#Very data types in DataFrame
covid_df.dtypes
```

```python
#Create a new csv file (cleaned) to import into analysis workbook
covid_df.to_csv("./Resources/country_vaccinations_clean.csv", index=False, header=True)
```
